```python
import numpy as np
import pandas as pd
```

```python
l=["google","facebook","yahoo","youtube","wikipedia","flipkart","amazon"]
n=7 #no. of sites
m=20 #no. of users
s=300 #no. of data
av_diff=[[(-1,-1) for _ in xrange(n)]for _ in xrange(n)]
d={}
a=[[-1 for _ in xrange(n)]for _ in xrange(m)]
r1=np.random.randint(0,n,s) #generating random sites
r2=np.random.randint(0,m,s) #generating random users
r3=[(r2[i],r1[i]) for i in xrange(s)] #merging users and sites
ind2=["User "+str(i) for i in xrange(m)]
```

```python
for i in xrange(s):
    user="User "+str(r3[i][0])
    site=l[r3[i][1]]
    key=user,site
    tmp=a[r3[i][0]][r3[i][1]]
    if(tmp==-1):
        a[r3[i][0]][r3[i][1]]=1
    else:
        a[r3[i][0]][r3[i][1]]+=1
    if key in d:
        d[key]+=1
    else:
        d[key]=1
```

```python
df1=pd.DataFrame(a,index=ind2,columns=l)
df1
```

```python
#Missing Ratings
df3=pd.DataFrame()
# miss_rat=df1[df1.google==-1]
miss_ind=[]
for i in l:
    miss_ind.append(df1[i]==-1)
    df2=df1.loc[df1[i]==-1]
    df3=df3.append(df2)
grouped = df3.groupby(level=0)
df4 = grouped.last()
df4
```

```python
#Constructing the item similarity matrix
for i in xrange(n):
    for j in xrange(n):
        tmp=0.0
        cnt=0.0
        for k in xrange(m):
            if(a[k][i]==-1 or a[k][j]==-1):
                continue
            tmp+=(a[k][i]-a[k][j])
            cnt+=1
        tmp_av=round(tmp/cnt,2)
        av_diff[i][j]=(tmp_av,cnt) #Storing pair(average difference,no. of users who rated both i & j site)
```

```python
df2=pd.DataFrame(av_diff,index=l,columns=l)
df2 #Item similarity matrix
```

```python
# Filling missing ratings by taking weighted mean
for i in xrange(m):
    for j in xrange(n):
        if(a[i][j]!=-1):continue #skipping those sites which are already rated
        num=0.0
        den=0.0
        for k in xrange(n):
            if(a[i][k]==-1):continue #skipping those site which are not rated
            num+=(av_diff[j][k][0]+a[i][k])*av_diff[j][k][1] #Weighted sum
            den+=av_diff[j][k][1]
        a[i][j]=int(num/den)
```

```python
df3=pd.DataFrame(a,index=ind2,columns=l)
df3
```
